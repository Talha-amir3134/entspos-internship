```python
class BankAccount:

    bank_name = "PyBank"
    total_accounts = 0


    #Constructor
    def __init__(self, owner, balance=0):

        self.owner = owner            #Public
        self._currency = "USD"        #Protected
        self.__balance = 0            #Private

        BankAccount.total_accounts += 1

        if balance > 0:
            self.deposit(balance)

    @property
    def balance(self):
        return self.__balance

    @balance.setter
    def balance(self, value):

        if value < 0:
            print("Balance cannot be negative")
        else:
            self.__balance = value


    #Instance method
    def deposit(self, amount):

        if amount <= 0:
            print("Invalid deposit amount")
            return

        self.__balance += amount
        print("Deposited:", amount)


    #Instance method
    def withdraw(self, amount):

        if amount <= 0:
            print("Invalid withdrawal amount")
            return

        if amount > self.__balance:
            print("Insufficient balance")
            return

        self.__balance -= amount
        print("Withdrawn:", amount)


    #Class method
    @classmethod
    def bank_info(cls):

        return "Bank:", cls.bank_name, "| Total accounts:", cls.total_accounts


    #Static method
    @staticmethod
    def is_valid_amount(amount):

        return isinstance(amount, (int, float)) and amount > 0


    #Readable output
    def __str__(self):

        return f"{self.owner}'s Account | Balance: ${self.__balance}"


    #Developer output
    def __repr__(self):

        return f"BankAccount(owner='{self.owner}', balance={self.__balance})"
```

```python
talha = BankAccount("talha", 1000)
arslan = BankAccount("Arslan", 500)

print("\nAccounts")
print(talha)
print(arslan)


print("\nDeposits")
talha.deposit(200)
```

```python
print("\nWithdrawals")
arslan.withdraw(100)


print("\nBalances using property")
print("talha balance:", talha.balance)
print("Arslan balance:", arslan.balance)
```

```python
print("\nBank info")
print(BankAccount.bank_info())


print("\nStatic method example")
print(BankAccount.is_valid_amount(100))
print(BankAccount.is_valid_amount(-10))


print("\nrepr output")
print(repr(talha))
print(repr(arslan))
```
