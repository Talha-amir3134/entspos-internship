```python
Task1 :
Product Catalogue Cleaner: A supplier sends a raw product list with mixed types, 
duplicates, and missing values. Clean the data, convert all prices to float, group products 
into a dictionary by category, and print a neat summary
```

```python
raw_products = [
    ("Laptop", "Electronics", "1200"),
    ("Mouse", "Electronics", 25),
    ("Chair", "Furniture", "85.5"),
    ("Desk", "Furniture", None),
    ("Laptop", "Electronics", "1200"),  # duplicate
    ("Pen", None, "2"),
    ("Monitor", "Electronics", "invalid"),
]

clean_product_list = []

product_set = set()


#cleaning data
for name,category,price in raw_products:

    #handle invalid name
    if not isinstance(name,str):
        name = "unknown"

    #handle invalid category
    if not isinstance(category, str):
        category = "unclassified"

    #hande invalid price        
    try:
        price = float(price)
    except(TypeError,ValueError):
        price = 0.0

    product = (name,category,price)

    if product not in product_set:
        product_set.add(product)
        clean_product_list.append(product)


#grouping data

grouped_product_dict = {}
for product in clean_product_list:

    if product[1] not in grouped_product_dict:
        grouped_product_dict[product[1]] = []
    grouped_product_dict[product[1]].append((product[0],product[2]))

#summary

print("*"*8,"summary","*"*8)
for key, value in grouped_product_dict.items():
    print(f"{key}:   {value}")



```
