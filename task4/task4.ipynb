 E-Commerce Inventory Analyzer: A shop owner has a messy product list with 
duplicates, missing prices, and mixed categories. Clean the data, group by category, find the top 5 most expensive items per category, and produce a summary dictionary — 
written in a Pythonic style using comprehensions throughout. 

```python
raw_products = [
    ("Laptop", "Electronics", "1200"),
    ("Mouse", "Electronics", 25),
    ("Chair", "Furniture", "85.5"),
    ("Desk", "Furniture", None),
    ("Laptop", "Electronics", "1200"),   # duplicate
    ("Pen", None, "2"),
    ("Monitor", "Electronics", "invalid"),
    ("Table", "Furniture", "150")
]

clean_products = []

for name, category, price in raw_products:
    if category is None:
        continue

    try:
        price = float(price)
    except:
        continue

    clean_products.append((name, category, price))

unique_products = list(set(clean_products))

grouped = {}

for name, category, price in unique_products:

    if category not in grouped:
        grouped[category] = []

    grouped[category].append((name, price))

top_products = {}

for category, items in grouped.items():
    sorted_items = sorted(items, key=lambda x: x[1], reverse=True)
    top_products[category] = sorted_items[:5]

summary = {
    category: {
        "count": len(items),
        "most_expensive": items[0]
    }
    for category, items in top_products.items()
}

print("Top Products Per Category")

for category, items in top_products.items():
    print(f"\n{category}")
    for name, price in items:
        print(name, price)

print("\nSummary")
print(summary)
```

HR Employee Ranker: HR wants employees ranked by performance score, then by 
experience as a tiebreaker, then by name alphabetically. The system must also support 
quickly searching for a specific employee by name from the sorted list. 

```python
employees = [
    ("Alice", 90, 5),
    ("Bob", 90, 3),
    ("Charlie", 85, 4),
    ("David", 90, 5),
    ("Eve", 70, 2)
]


employees = sorted(employees, key=lambda e: e[0])

employees = sorted(employees, key=lambda e: e[2], reverse=True)

ranked_employees = sorted(employees, key=lambda e: e[1], reverse=True)

print("Employee Ranking")

for rank, emp in enumerate(ranked_employees, start=1):
    name, score, experience = emp
    print(f"{rank}. {name} | Score: {score} | Experience: {experience} years")

employees_by_name = sorted(ranked_employees, key=lambda x: x[0])

def binary_search_employee(arr, target):

    left = 0
    right = len(arr) - 1

    while left <= right:

        mid = (left + right) // 2
        name = arr[mid][0]

        if name == target:
            return arr[mid]

        elif name < target:
            left = mid + 1

        else:
            right = mid - 1

    return None


result = binary_search_employee(employees_by_name, "Charlie")

print("\nSearch Result:")
print(result)
```
